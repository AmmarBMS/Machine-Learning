```python
import pandas as pd
from google.colab import files
from sklearn.preprocessing import MinMaxScaler, StandardScaler


uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df_adult = pd.read_csv(file_name)

print("\nProcessing Dataset: Adult Dataset")


print("\nColumn Information:")
print(df_adult.info())


print("\nStatistical Information:")
print(df_adult.describe())


if "Ocean Proximity" in df_adult.columns:
    print("\nUnique counts in 'Ocean Proximity':")
    print(df_adult["Ocean Proximity"].value_counts())


missing_values = df_adult.isnull().sum()
missing_columns = missing_values[missing_values > 0]

print("\nColumns with missing values:")
print(missing_columns)


df_adult.fillna(df_adult.select_dtypes(include=['number']).mean(), inplace=True)
df_adult.fillna(df_adult.select_dtypes(include=['object']).mode().iloc[0], inplace=True)


categorical_columns = df_adult.select_dtypes(include=['object']).columns
print("\nCategorical columns identified for encoding:")
print(categorical_columns)

df_adult = pd.get_dummies(df_adult, columns=categorical_columns, drop_first=True)


num_cols = df_adult.select_dtypes(include=['number']).columns
Q1 = df_adult[num_cols].quantile(0.25)
Q3 = df_adult[num_cols].quantile(0.75)
IQR = Q3 - Q1


df_adult = df_adult[~((df_adult[num_cols] < (Q1 - 1.5 * IQR)) | (df_adult[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]


min_max_scaler = MinMaxScaler()
df_scaled_minmax = pd.DataFrame(min_max_scaler.fit_transform(df_adult), columns=df_adult.columns)

standard_scaler = StandardScaler()
df_scaled_standard = pd.DataFrame(standard_scaler.fit_transform(df_adult), columns=df_adult.columns)


print("\nDifference between Min-Max Scaling and Standardization:")
print("""
- Min-Max Scaling (Normalization):
  - Rescales values to a range of 0 to 1.
  - Preserves original distribution.
  - Used when features have different scales but no extreme outliers.

- Standardization (Z-score normalization):
  - Centers data around mean (0) with a standard deviation of 1.
  - More robust to outliers.
  - Preferred for models that assume normally distributed data (e.g., PCA, linear regression).
""")

print("\nData processing complete for: Adult Dataset")
```

```python
import pandas as pd
from google.colab import files
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Upload dataset
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# Load the dataset
df_diabetes = pd.read_csv(file_name)

print("\nProcessing Dataset: Diabetes Dataset")

# Display column information
print("\nColumn Information:")
print(df_diabetes.info())

# Display statistical summary
print("\nStatistical Information:")
print(df_diabetes.describe())

# 1. Identify and handle missing values
missing_values = df_diabetes.isnull().sum()
missing_columns = missing_values[missing_values > 0]

print("\nColumns with missing values:")
print(missing_columns)

# Handling missing values:
# - Numerical columns: Fill with mean
# - Categorical columns: Fill with mode (most frequent value)
df_diabetes.fillna(df_diabetes.select_dtypes(include=['number']).mean(), inplace=True)
df_diabetes.fillna(df_diabetes.select_dtypes(include=['object']).mode().iloc[0], inplace=True)

# 2. Identify categorical columns and apply One-Hot Encoding
categorical_columns = df_diabetes.select_dtypes(include=['object']).columns
print("\nCategorical columns identified for encoding:")
print(categorical_columns)

df_diabetes = pd.get_dummies(df_diabetes, columns=categorical_columns, drop_first=True)

# 3. Handling outliers using IQR method (only for numerical columns)
num_cols = df_diabetes.select_dtypes(include=['number']).columns
Q1 = df_diabetes[num_cols].quantile(0.25)
Q3 = df_diabetes[num_cols].quantile(0.75)
IQR = Q3 - Q1

# Removing outliers
df_diabetes = df_diabetes[~((df_diabetes[num_cols] < (Q1 - 1.5 * IQR)) | (df_diabetes[num_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

# 4. Apply Min-Max Scaling and Standardization
min_max_scaler = MinMaxScaler()
df_scaled_minmax = pd.DataFrame(min_max_scaler.fit_transform(df_diabetes), columns=df_diabetes.columns)

standard_scaler = StandardScaler()
df_scaled_standard = pd.DataFrame(standard_scaler.fit_transform(df_diabetes), columns=df_diabetes.columns)

# Explanation of Scaling Methods
print("\nDifference between Min-Max Scaling and Standardization:")
print("""
- Min-Max Scaling (Normalization):
  - Rescales values between 0 and 1.
  - Preserves distribution but sensitive to outliers.
  - Used when features have different scales without extreme outliers.

- Standardization (Z-score normalization):
  - Centers data around mean (0) with standard deviation of 1.
  - More robust to outliers.
  - Preferred for models that assume normal distribution (e.g., PCA, regression).
""")

print("\nData processing complete for: Diabetes Dataset")
```
